1. Eficiencia en el tiempo de ejecución: Mientras que la fuerza bruta explora todas las combinaciones posibles de soluciones, lo que lleva a una complejidad exponencial, GRASP utiliza una heurística para generar soluciones iniciales aleatorias y luego aplica una búsqueda local para mejorarlas. Esta estrategia permite obtener soluciones de alta calidad en tiempos de ejecución más razonables.

2. Escalabilidad: A medida que el tamaño del problema aumenta, la fuerza bruta se vuelve impracticable debido a la explosión combinatoria. En cambio, GRASP, al ser una heurística, puede manejar problemas más grandes con tiempos de ejecución razonables. Esto hace que GRASP sea escalable y aplicable a casos prácticos con mayor cantidad de datos.

3. Adaptabilidad: GRASP es un algoritmo flexible que se puede ajustar mediante la selección de parámetros, como el número de iteraciones de búsqueda local o el nivel de aleatoriedad en la generación de soluciones iniciales. Esta característica permite adaptar GRASP a diferentes tipos de problemas y obtener soluciones adecuadas para cada situación particular.

La función objetivo en este problema del doblaje de la película es la meta que buscamos alcanzar, y consiste en **minimizar** el gasto total por los servicios de los actores de doblaje a lo largo de todo el proceso de grabación.

```
import pandas as pd
import itertools

df = pd.read_excel('/content/Datos problema doblaje(30 tomas, 10 actores) (2).xlsx', index_col=0, header=0)

# Convertir los datos en un DataFrame
df = pd.DataFrame(df, index=range(1, 31), columns=range(1, 11))

display(df.head())


def calcular_costo_combinacion(combination, df):
    """
    Calcula el costo de una combinación dada de tomas y actores.
    El costo se determina por la cantidad de días que los actores deben desplazarse al estudio de grabación.
    """
    dias_totales = 0
    for day in combination:
        actores_en_day = set(df.columns[i] for i in range(len(df.columns)) if day[i] == 1)
        dias_totales += len(actores_en_day) - 1  # Restar 1 para considerar que el primer actor no necesita desplazarse
    return dias_totales

def planificar_doblaje(df):
    actores = df.columns.tolist()
    tomas = df.index.tolist()

    mejor_combinacion = None
    menor_costo = float('inf')

    # Generar todas las combinaciones posibles de tomas y actores por día
    for combination in itertools.product([0, 1], repeat=len(tomas) * len(actores)):
        # Dividir la combinación en días
        dias = [combination[i:i + len(actores)] for i in range(0, len(combination), len(actores))]

        # Verificar que no haya más de 6 tomas por día
        if all(sum(day) <= 6 for day in dias):
            costo_combinacion = calcular_costo_combinacion(dias, df)
            if costo_combinacion < menor_costo:
                mejor_combinacion = dias
                menor_costo = costo_combinacion

    return mejor_combinacion, menor_costo

mejor_combinacion, costo_optimo = planificar_doblaje(df)
print("Mejor combinación de tomas por día:", mejor_combinacion)
print("Menor costo:", costo_optimo)

```

```
"""Algoritmo GRASP"""


def calcular_costo_dia(dia, df):
    """
    Calcula el costo de un día dado (una lista de 1's y 0's) en función de la cantidad de actores que deben desplazarse.
    """
    actores_en_dia = set(df.columns[i] for i in range(len(df.columns)) if dia[i] == 1)
    return len(actores_en_dia) - 1  # Restar 1 para considerar que el primer actor no necesita desplazarse

def generar_solucion_aleatoria(df):
    """
    Genera una solución aleatoria para la planificación de tomas.
    """
    tomas = df.index.tolist()
    solucion = [random.randint(0, 1) for _ in range(len(tomas) * len(df.columns))]
    dias = [solucion[i:i + len(df.columns)] for i in range(0, len(solucion), len(df.columns))]
    return dias


def mejorar_solucion(solucion, df):
    tomas = df.index.tolist()
    for _ in range(100):
        dia_idx = random.randint(0, len(solucion) - 1)
        dia_actual = solucion[dia_idx]

        tomas_en_dia = [i for i, toma in enumerate(dia_actual) if toma == 1]
        if not tomas_en_dia:
            continue  # Si no hay tomas en el día, omitir esta iteración

        toma_a_cambiar = random.choice(tomas_en_dia)
        nuevo_estado_toma = 1 - dia_actual[toma_a_cambiar]
        dia_actual[toma_a_cambiar] = nuevo_estado_toma

        costo_actual = calcular_costo_dia(dia_actual, df)
        if costo_actual > 6:
            dia_actual[toma_a_cambiar] = 1 - nuevo_estado_toma

def planificar_doblaje_grasp(df, max_iter=100):
    """
    Implementación de GRASP para planificar el doblaje de una película.
    """
    mejor_solucion = None
    menor_costo = float('inf')

    for _ in range(max_iter):
        solucion_inicial = generar_solucion_aleatoria(df)
        mejorar_solucion(solucion_inicial, df)

        costo_actual = sum(calcular_costo_dia(dia, df) for dia in solucion_inicial)
        if costo_actual < menor_costo:
            mejor_solucion = solucion_inicial
            menor_costo = costo_actual

    return mejor_solucion, menor_costo


mejor_solucion, costo_optimo = planificar_doblaje_grasp(df)
print("\nMejor planificación de tomas por día:")
display(mejor_solucion)
print("\nMenor costo:", costo_optimo)

```
